```python
#organizar los nombres de las columnas en un diccionario
# las que comiencen con "NHC", se ingresaran en una lista de columnas como valor con la letra inmediatamente contigua a "NHC" como llave. 
# Por ejemplo, para una columna llamada "NHCBP1", se ingresará al diccionario como "B": ["NHCBP1", "NHCBP2", "NHCBP3", ...]
# Para las columnas que comiencen con "NPC", se ingresaran en una lista de columnas como valor con la letra inmediatamente contigua a "NPC" como llave.
# Por ejemplo, para una columna llamada "NPCHP1", se ingresará al diccionario como "H": ["NHPP1", "NHPP2", "NHPP3", ...]
# Para las columnas que no cumplan ninguna de las condiciones anteriores, se ingresaran en una lista de columnas como valor con la letra "otros" como llave.
# Para las columnas que comiencen con "NVC", se ingresaran en una lista de columnas como valor con la letra inmediatamente contigua a "NVC" como llave.
# Por ejemplo, para una columna llamada "NVCHP1", se ingresará al diccionario como "H": ["NVCHP1", "NVCHP2", "NVCHP3", ...]
def organizar_columnas(df):
    columnas = df.columns
    columnas_dict = {}
    for columna in columnas:
        if columna.startswith("NHC"):
            letra = columna[3]
            if letra in columnas_dict:
                columnas_dict[letra].append(columna)
            else:
                columnas_dict[letra] = [columna]
        elif columna.startswith("NPC"):
            letra = columna[3]
            if letra in columnas_dict:
                columnas_dict[letra].append(columna)
            else:
                columnas_dict[letra] = [columna]
        elif columna.startswith("NVC"):
            letra = columna[3]
            if letra in columnas_dict:
                columnas_dict[letra].append(columna)
            else:
                columnas_dict[letra] = [columna]
        else:
            if "otros" in columnas_dict:
                columnas_dict["otros"].append(columna)
            else:
                columnas_dict["otros"] = [columna]
    return columnas_dict
```

```python
# Aplicamos la función a tus DataFrames
columnas_ordenadas21 = organizar_columnas(archivo21_filtrado)
columnas_ordenadas17 = organizar_columnas(archivo17_filtrado)

```

```python
# Organizar la información de los diccionarios. Cada pregunta tiene el formato "NPCFP1A", donde los tres primeros digitos
# representan un ID, la letra continua la categoría, luego "P#" representa el numero de la pregunta y las letras continuas
# (en caso de que las haya porque no siempre hay) representan las subpreguntas. 
# Se debe organizar la información de tal manera que se pueda acceder a la información de cada pregunta general de manera sencilla.
# Diccionario de preguntas cuyos valores son subpreguntas

def organize_questions(dictionary):
    # Create a dictionary to store the questions
    questions = {}
    
    # Iterate over the keys in the input dictionary
    for key in dictionary:
        # Extract the digits from the key
        digits = ''.join(filter(str.isdigit, key))
        
        # Initialize question_id with a default value
        question_id = digits

        # Add the question to the dictionary if it doesn't exist
        if question_id is not None and question_id not in questions:
            questions[question_id] = dictionary[key]
    
    return questions


```

```python
# Imprimir las preguntas de cada sección
def print_questions(questions):
    # Iterate over the questions
    for question in questions:
        # Print the question followed by a new line
        print(question, '\n')

for i in columnas_ordenadas21:
    if i != "otros":
        print_questions(organize_questions(columnas_ordenadas21[i]))
```
